```python
#Import the necessary packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
%matplotlib inline
```

```python
#Read the input file in variable facebook

path = 'C:/Users/Vasu/Desktop/Vasu/LEARNING/data Analytics/UPX/Projects/Project5_Facebook/pseudo_facebook.tsv'
fb = pd.read_csv(path,sep='\t')
fb.tail()
```

```python
#facebook variable no of rows and columns
fb.shape
```

```python
# setting index of te dataframe as userid

fb = fb.set_index('userid')
fb.head()
```

```python
#Zodiac Sign - Date Range
#Aquarius (January 20 - February 18) 
#Pisces (February 19 - March 20) 
#Aries (March 21 - April 19) 
#Taurus (April 20 - May 20) 
#Gemini (May 21 - June 20) 
#Cancer (June 21 - July 22) 
#leo (July 23 - August 22) 
#Virgo (August 23 - September 22) 
#Libra (September 23 - October 22) 
#Scorpio (October 23 - November 21) 
#Sagittarius (November 22 - December 21) 
#Capricorn (December 22 - January 19) 

# Define Function
def Zodiac_Sign (row):
    if (row['dob_month']==11 and row['dob_day'] >= 22) or (row['dob_month']==12 and row['dob_day'] <= 21):
       return 'SAGITTARIUS'
    if (row['dob_month']==10 and row['dob_day'] >= 23) or (row['dob_month']==11 and row['dob_day'] <= 21):
       return 'SCORPIO'
    if (row['dob_month']==9 and row['dob_day'] >= 23) or (row['dob_month']==10 and row['dob_day'] <= 22):
       return 'LIBRA'
    if (row['dob_month']==8 and row['dob_day'] >= 23) or (row['dob_month']==9 and row['dob_day'] <= 22):
       return 'VIRGO'
    if (row['dob_month']==7 and row['dob_day'] >= 23) or (row['dob_month']==8 and row['dob_day'] <= 22):
       return 'LEO'
    if (row['dob_month']==6 and row['dob_day'] >= 21) or (row['dob_month']==7 and row['dob_day'] <= 22):
       return 'CANCER'
    if (row['dob_month']==5 and row['dob_day'] >= 21) or (row['dob_month']==6 and row['dob_day'] <= 20):
       return 'GEMINI'
    if (row['dob_month']==4 and row['dob_day'] >= 20) or (row['dob_month']==5 and row['dob_day'] <= 20):
       return 'TAURAS'
    if (row['dob_month']==3 and row['dob_day'] >= 21) or (row['dob_month']==4 and row['dob_day'] <= 19):
       return 'ARIES'
    if (row['dob_month']==2 and row['dob_day'] >= 19) or (row['dob_month']==3 and row['dob_day'] <= 20):
       return 'PISCES'
    if (row['dob_month']==1 and row['dob_day'] >= 20) or (row['dob_month']==2 and row['dob_day'] <= 18):
       return 'AQUARIUS'
    return 'CAPRICON'
```

```python
fb['Zodiac_sign'] = fb.apply (lambda row: Zodiac_Sign(row),axis=1)
fb['Zodiac_sign']
fb.head()
```

```python
#Use describe function to see the various aspects of facebbok data

fb.describe()
```

```python
#To Cater Average Likes_Received and Average Friends Count for all users

fb_all = fb[['Zodiac_sign','friend_count','likes_received']]
fb_all = fb_all.groupby('Zodiac_sign').mean().round()
fb_all.head(2)
```

```python
#To Cater Average Likes_Received and Average Friends Count for Females

fb_females = fb[fb['gender']=='female']
fb_females = fb_females[['Zodiac_sign','friend_count','likes_received']]
fb_females = fb_females.groupby('Zodiac_sign').mean().round()
fb_females.head(2)
```

```python
#To Cater Average Likes_Received and Average Friends Count for Males

fb_males = fb[fb['gender']=='male']
fb_males = fb_males[['Zodiac_sign','friend_count','likes_received']]
fb_males = fb_males.groupby('Zodiac_sign').mean().round()
fb_males.head(2)
```

```python
#Restting the index for all the 2 dataframes.

fb_all = fb_all.reset_index()
fb_males = fb_males.reset_index()
fb_females = fb_females.reset_index()
```

```python
fb_females.head(2)
fb_males.head(2)
```

```python
#Renaming the columns of the Dataframes

fb_all.columns = ['Zodiac_sign','All_friend_count','All_likes_received']
fb_males.columns = ['Zodiac_sign','Males_friend_count','Males_likes_received']
fb_females.columns = ['Zodiac_sign','Females_friend_count','Females_likes_received']
fb_females.head(2)
```

```python
#Concatinating the 3 dataframe sinto one.

fb_count = pd.concat([fb_all,fb_males,fb_females],axis=1)
fb_count
```

```python
#Lets plot bar graph for 'Avg Friend Count' & 'Avg Likes_received' Vs Zodiac Sign

n_groups = 12
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = .85

axis_font = {'fontname':'Arial', 'size':'14'}
title_font = {'fontname':'Arial', 'size':'16', 'color':'blue'}

rects1 = plt.bar(index, fb_count['All_friend_count'], bar_width,
                 alpha=opacity,
                 color='green',
                 label='Avg. Friend Count')
 
rects2 = plt.bar(index + bar_width, fb_count['All_likes_received'], bar_width,
                 alpha=opacity,
                 color='red',
                 label='Avg.Likes_Received',)
 
plt.xlabel('Zodiac Sign',axis_font)
plt.title('Avg number of Likes_Received and Friend Count/Zodiac Sign',axis_font)
plt1 = plt.xticks(index + bar_width, ('AQUARIES','ARIES','CANCER','CAPRICON','GEMINI','LEO','LIBRA','PISCIES','SAGITTARIUS','SCORPIO','TAURAS','VIRGO'
                    ))
fig = plt.gcf()
fig.set_size_inches(15.5, 8.5)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Lets plot bar graph for 'Friend_Count Vs Zodiac Sign' for Males/Females


n_groups = 12
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = .85

axis_font = {'fontname':'Arial', 'size':'14'}
title_font = {'fontname':'Arial', 'size':'16', 'color':'blue'}

rects1 = plt.bar(index, fb_count['Males_friend_count'], bar_width,
                 alpha=opacity,
                 color='green',
                 label='Avg. Male Friend Count')
 
rects2 = plt.bar(index + bar_width, fb_count['Females_friend_count'], bar_width,
                 alpha=opacity,
                 color='red',
                 label='Avg.Female Friend Count',)
 
plt.xlabel('Zodiac Sign',axis_font)
plt.ylabel('Avg. Friend_Count',axis_font)
plt.title('Friend_Count Vs Zodiac Sign for Males/Females',axis_font)
plt1 = plt.xticks(index + bar_width, ('AQUARIES','ARIES','CANCER','CAPRICON','GEMINI','LEO','LIBRA','PISCIES','SAGITTARIUS','SCORPIO','TAURAS','VIRGO'
                                      ))
fig = plt.gcf()
fig.set_size_inches(15.5, 8.5)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Lets plot bar graph for 'Likes_Received Vs Zodiac Sign' for Males/Females

n_groups = 12
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = .85

axis_font = {'fontname':'Arial', 'size':'14'}
title_font = {'fontname':'Arial', 'size':'16', 'color':'blue'}

rects1 = plt.bar(index, fb_count['Males_likes_received'], bar_width,
                 alpha=opacity,
                 color='green',
                 label='Avg. Male Likes_Received')
 
rects2 = plt.bar(index + bar_width, fb_count['Females_likes_received'], bar_width,
                 alpha=opacity,
                 color='red',
                 label='Avg.Female Likes_Received',)
 
plt.xlabel('Zodiac Sign',axis_font)
plt.ylabel('Avg. Likes_Received',axis_font)
plt.title('Likes_Received Vs Zodiac Sign for Males/Females',axis_font)
plt1 = plt.xticks(index + bar_width, ('AQUARIES','ARIES','CANCER','CAPRICON','GEMINI','LEO','LIBRA','PISCIES','SAGITTARIUS','SCORPIO','TAURAS','VIRGO'
                                      ))
fig = plt.gcf()
fig.set_size_inches(15.5, 8.5)
plt.legend()

plt.tight_layout()
plt.show()
```
