```python
#Import the necessary packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
%matplotlib inline
```

```python
#Read the input file in variable facebook

path = 'C:/Users/Vasu/Desktop/Vasu/LEARNING/data Analytics/UPX/Projects/Project5_Facebook/pseudo_facebook.tsv'
fb = pd.read_csv(path,sep='\t')
fb.tail()
```

```python
#facebook variable no of rows and columns
fb.shape
```

```python
# setting index of te dataframe as userid

fb = fb.set_index('userid')
fb.head()
```

```python
#Use describe function to see the various aspects of facebbok data

fb.describe()
```

```python
#Lets try for Q07, so see if there is any relation between Month of Birth Vs Likes Received

fb_males = fb[fb['gender']=='male']
fb_males = fb_males[['dob_month','likes_received']]
fb_females = fb[fb['gender']=='female']
fb_females = fb_females[['dob_month','likes_received']]
fb_females = fb_females.groupby('dob_month').mean()
fb_females
fb_males = fb_males.groupby('dob_month').mean()
fb_males
```

```python
fb_females
```

```python
fb_likes_month = [fb_males,fb_females]
result = pd.concat(fb_likes_month,axis=1)
result
result=result.reset_index()
```

```python
result.columns = ['month','males','females']
result
```

```python
#Lets plot bar graph for 'Month of Birth' Vs Likes Received

n_groups = 12
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = .85
 
rects1 = plt.bar(index, result['males'], bar_width,
                 alpha=opacity,
                 color='blue',
                 label='Males')
 
rects2 = plt.bar(index + bar_width, result['females'], bar_width,
                 alpha=opacity,
                 color='pink',
                 label='Females')
 
plt.xlabel('Month')
plt.ylabel('Likes Received')
plt.title('Month of Birth Vs Likes Received')
plt.xticks(index + bar_width, ('Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'))
plt.legend()
 
plt.tight_layout()
plt.show()
```

```python
#Lets also try the same plot in Line graph as well to compare if the above plot looks much better in Line graph or not.

plt.plot(result['month'],result['males'],'blue',linewidth=4.0)
plt.plot(result['month'],result['females'],'green',linewidth=4.0)
plt.xlabel('Month of Birth',fontsize = 'x-large')
plt.ylabel('Avg no of Friends',fontsize = 'x-large')
plt.legend(['males','females'],loc='upper left',fontsize = 'x-large')
plt.show()
```
